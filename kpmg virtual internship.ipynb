### Categorical Univariate Analysis

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(data=df_new,x='product_class',order=df_new['product_class'].value_counts().index)
total = float(len(df))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format((height/total)*100),
            ha="center")
plt.show()
```
